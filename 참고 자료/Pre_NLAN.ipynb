```python
with open('익스포트.csv') as file:
    csv_data = []
    for line in file.readlines():
        csv_data.append(line.split(','))

for i in range (1, 4999):
    print(csv_data[i][2], '|' , csv_data[i][3])
```

```python
from keras.preprocessing.text import Tokenizer
t = Tokenizer()
t.fit_on_texts(csv_data)
print(t.word_index)
```
